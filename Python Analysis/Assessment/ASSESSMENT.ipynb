```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cardiovascular dataset.csv",sep=';')
df
```

```python
# 1. What is the shape of the dataset?
print(f"Rows in dataset are : {df.shape[0]}")
print(f"Columns in dataset are : {df.shape[1]}")
print(f"Shape is : {df.shape}")
```

```python
# 2. Are there any missing values in the dataset?
print(f"Missing values in dataset are : {df.isnull().sum()}")
print("")
print("No missing values")
```

```python
# 3. What are the unique values for categorical features like gender, cholesterol, gluc?
print(f"Unique values in Gender are : {df['gender'].unique()}  \nAnd the count for it is as :\n{df['gender'].value_counts()}\n")
print(f"Unique values in Gender are : {df['cholesterol'].unique()}  \nAnd the count for it is as :\n{df['cholesterol'].value_counts()}\n")
print(f"Unique values in Gender are : {df['gluc'].unique()}  \nAnd the count for it is as :\n{df['gluc'].value_counts()}")
```

```python
# 4. What’s the average age of patients (in years)?
m = df['age'].mean()
print(f"Average age of the patients is : {m:.0f} years.")
```

```python
# 5. What is the distribution of the target variable (cardio)?
print(df['cardio'].value_counts())
z = df[df['cardio']==0]
o = df[df['cardio']==1]
plt.figure(figsize=(6,4))
plt.title(f"Line chart for zeros")
for cols in z:
    sns.lineplot(data=df['id'],color='red')
plt.show()
plt.figure(figsize=(6,4))
plt.title(f"Line chart for zeros")
for cols in o:
    sns.lineplot(data=df['id'],color='blue')
plt.show()
```

```python
#10. How many rows have incorrect blood pressure values?
print(f"The cases where BP is over 120 are: {df[(df['ap_hi']>120)].shape[0]}")
# df[(df['ap_hi']>120)]

print(f"The cases where BP is less than 80 are: {df[(df['ap_lo']<80)].shape[0]}")
# df[(df['ap_lo']<80)]
```

```python
# 11. Remove rows with invalid blood pressure, height, or weight?
df = df[~((df['height'] > 220) | (df['height'] < 120))]      #120-220
df = df[~((df['weight']<40) | (df['weight']>200))]             # 40-200
df = df[~(
    (df['ap_lo']<80) | (df['ap_lo']>120) |
    (df['ap_hi']<80) | (df['ap_hi']>120)
)]
```

```python
# 17. Plot age distribution for those with and without cardio disease
fig,ax = plt.subplots(nrows=1,ncols=2,sharey=True)
counts_is = is_cardio['age'].value_counts()
counts_not = not_cardio['age'].value_counts()
ax[0].scatter(x=counts_is.index[:50],y=counts_is.index[:50])
ax[1].scatter(x=counts_not.index[:50],y=counts_not.index[:50],color='red')
```

```python
# 20. What percentage of patients have above-normal glucose levels?

total = df['gluc'].value_counts().sum()
# 1 → Normal
# 2 → Above normal (prediabetic / slightly high)
# 3 → Well above normal (high / diabetic range

above_n = df[df['gluc']==2]
s = above_n['gluc'].sum()          #3086
val = int(s)
per = (val/total)*100
print(f"Percentage of patients having above normal glucose levels are : {per:.2f}")
```
